```python
import pandas as pd
```

```python
data = pd.read_csv("news_collection21.csv")
```

```python
data["title"].head()
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
content_list = []
c=0
for i in range(len(data)):
    c+=1
    content_dict = {}
    title = data["title"].iloc[i]
    doc = nlp(title)
    name_entity_list=[]
    for ent in doc.ents:   
        name_entity = {}
        name_entity["entities"] = ent
        name_entity["labels"]   = ent.label_
        name_entity_list.append(name_entity)
        
    content_dict["title"] = title
    content_dict["name entity recognition"] = name_entity_list
    content_list.append(content_dict)
    if c==500:
        break
    
    
```

```python
content_list
```

```python
df = pd.DataFrame.from_dict(content_list, orient='columns') 
df.to_excel("name_entity_result.xlsx",engine='xlsxwriter')  
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```

```python
ex = 'European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices'
```

```python
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
```

```python
sent = preprocess(ex)
sent
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
```

```python
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs)
```

```python
from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint
iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "My name is Wolfgang and I live in Berlin"

ner_results = nlp(example)
print(ner_results)
```
